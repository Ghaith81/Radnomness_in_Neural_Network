```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
tf.keras.backend.set_floatx('float64')
np.set_printoptions(suppress=True)


def xor(gradient_noise, loss_noise, optimizer, verbose=0, plot=0):


    tf.compat.v1.disable_eager_execution()  # Disable eager execution

    # Define input data
    x1_data = np.array([[1], [0], [1], [0]], dtype=np.float32)
    x2_data = np.array([[0], [1], [1], [0]], dtype=np.float32)
    y_true_data = np.array([[1], [1], [1], [0]], dtype=np.float32)

    # Define the computation graph
    W = tf.Variable(tf.random.normal([2, 1]), name='weights')
    b = tf.Variable(tf.zeros([1]), name='bias')

    x = tf.concat([x1_data, x2_data], axis=1)
    logits = tf.matmul(x, W) + b
    y_pred = tf.sigmoid(logits)


    # Define the loss function
    l = tf.keras.losses.binary_crossentropy(y_true_data, y_pred)
    loss = l * tf.random.normal(tf.shape(l), 1.0, loss_noise)

    # Define the optimizer and training operation with gradient noise
    learning_rate = 0.1
    optimizer = optimizer(learning_rate=learning_rate)

    grads_and_vars = optimizer.compute_gradients(loss, var_list=[W, b])
    grads = [grad for grad, var in grads_and_vars]
    t = tf.Variable(0, dtype=tf.float32)
    gradient_noise = tf.constant(gradient_noise)
    gradient_noise_std = tf.math.sqrt(tf.math.divide_no_nan(tf.cast(gradient_noise, tf.float32), tf.math.pow(t+1.0, 0.55)))
    noisy_grads = [grad + tf.random.normal(grad.shape, mean=0.0, stddev=gradient_noise_std) for grad in grads]
    train_op = optimizer.apply_gradients(zip(noisy_grads, [W, b]))
    t_assign = tf.compat.v1.assign_add(t, 1.0)


    # Create a session and initialize variables
    sess = tf.compat.v1.Session()
    sess.run(tf.compat.v1.global_variables_initializer())

    # Train the model and store W, b, grad_W, and grad_b values at each step
    w_values = []
    b_values = []
    grad_w_values = []
    grad_b_values = []
    for i in range(1000):
        # Run a training step with gradient noise
        _, loss_value, W_value, b_value, grad_W_value, grad_b_value = sess.run([train_op, loss, W, b, grads[0], grads[1]])

        w_values.append(W_value)  # Store W value
        b_values.append(b_value)  # Store b value
        grad_w_values.append(grad_W_value)  # Store grad_W value
        grad_b_values.append(grad_b_value)  # Store grad_b value

        sess.run(t_assign)  # Increment time step counter

        if i % 100 == 0 and verbose:
            print("Step:", i, "Loss:", loss_value)
    
    if plot:

        # Test the model on new data
        y_pred_test = sess.run(y_pred)
        #print("Predictions:", y_pred_test)

        # Convert parameter and gradient lists to NumPy arrays
        w_values = np.array(w_values)
        b_values = np.array(b_values)
        grad_w_values = np.array(grad_w_values)
        grad_b_values = np.array(grad_b_values)

        # Save parameter and gradient arrays to disk
        np.save("w_values.npy", w_values)
        np.save("b_values.npy", b_values)
        np.save("grad_w_values.npy", grad_w_values)
        np.save("grad_b_values.npy", grad_b_values)

        # Plot the parameter values
        # Plot W and b values over training steps
        plt.plot(w_values[:, 0], label='W1')
        #plt.plot(w_values[:, 1], label='W2')
        #plt.plot(b_values[:, 0], label='b')
        plt.xlabel('Training step')
        plt.ylabel('Parameter value')
        #plt.legend()
        plt.show()

        # Plot grad_W and grad_b values over training steps
        plt.plot(grad_w_values[:, 0], label='dL/dW1')
        #plt.plot(grad_w_values[:, 1], label='dL/dW2')
        #plt.plot(grad_b_values[:, 0], label='dL/db')
        #plt.ylim(-0.2, 0.0)
        plt.xlabel('Training step')
        plt.ylabel('Gradient value')
        #plt.legend()
        plt.show()


        #plt.hist(grad_w_values[:, 0], bins=50)
        #plt.show()
        
    return (sess.run(tf.reduce_mean(tf.keras.losses.binary_crossentropy(y_true_data, y_pred))))

```

```python
import pandas as pd
import seaborn as sns

df = pd.DataFrame(columns=['method', 'loss'])

optimizer = tf.compat.v1.train.GradientDescentOptimizer

optimizer = tf.compat.v1.train.AdamOptimizer

while True:
    loss = xor(0.5, 0.0, optimizer, 1, 1)
    df.loc[len(df)] = ['gradient', loss]
    
    loss = xor(0.0, 0.5, optimizer, 0, 1)
    df.loc[len(df)] = ['loss', loss]
    
    loss = xor(0.0, 0.0, optimizer, 0, 1)
    df.loc[len(df)] = ['None', loss]
    
    sns.boxplot(x='method', y = 'loss', data=df)
    plt.show()
    
    

```
