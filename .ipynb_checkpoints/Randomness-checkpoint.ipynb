```python
from NeuralNetwork import NeuralNetwork, VGG1
from Dataset import Dataset, SPIRAL, MOON, CIRCLE
from NeuroEvolution import NeuroEvolution
from tensorflow import keras
from Representation import Function, PAU
from Activation import PANGAEA_Activation, Pade_Activation_Unit
import tensorflow as tf
import numpy as np
import pandas as pd
import time
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
num_classes = 10
input_shape = (28, 28, 1)
mnist_dataset = Dataset(X_train, y_train, X_test, y_test, num_classes, input_shape, expand=True, label_noise=0)

model_vgg1 = VGG1(mnist_dataset)
```

```python
loss = Function([1, 15, 26, 1, 3, 9, 30, 8, 17])
activation =  PANGAEA_Activation([6, 19, 26, 3, 5, 9, 19, 22, 28])
#activation =  Pade_Activation_Unit()

gradient = Function([1, 31, 31, 1, 1, 31, 31, 31, 31])
optimizer = tf.optimizers.Adam()
dropout = 0.5
metric = keras.metrics.CategoricalAccuracy()
batch_size = 128
epochs = 3
patience = np.inf 
verbose = 1


model_vgg1.set_config(loss, activation, gradient, optimizer, dropout, metric, batch_size, epochs, patience, verbose=verbose)
model_vgg1.create_model()
```

```python
model_vgg1.fit()
```

```python
model_vgg1.test_score
```
