```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
num_classes = 10
input_shape = (28, 28, 1)
mnist_dataset = Dataset(X_train, y_train, X_test, y_test, num_classes, input_shape, expand=True, label_noise=0)

model_vgg1 = VGG1(mnist_dataset)
```

```python
def plotPAU(f):
    plotRange = np.linspace(-2, 2,1000)

    values = []
    for i in plotRange:
                    values.append(f.calculate(float(i)))
    plt.plot(plotRange, values)
    plt.xlim(-2,2)
    #plt.ylim(-4,4)
    plt.show()

def plotFunction(f):
    plotRange = np.linspace(-1, 1,1000)

    values = []
    for i in plotRange:
                    values.append(f.s2(float(i)))
    plt.plot(plotRange, values)
    plt.xlim(-1,1)
    #plt.ylim(-4,4)
    plt.show()
```
