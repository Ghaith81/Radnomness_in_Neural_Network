```python
import matplotlib.pyplot as plt
from MNIST import MNIST, MNISTN1, SPIRAL
from Representation import Function
import tensorflow.keras as keras
from NeuroEvolution import NeuroEvolution
import numpy as np
from Models.Models import CIFAR10, CIFAR10WRN
```

```python
def plotFunction(f):
    plotRange = np.linspace(-20, 20,1000)

    values = []
    for i in plotRange:
                    values.append(f.s2(float(i)))
    plt.plot(plotRange, values)
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    plt.show()
```

# ReLU

```python
f = Function([4, 15, 10, 1, 4, 16, 25, 11, 12])
model = SPIRAL(1000, 1, f)
```

```python
model.visualizeTrainTest()
```

```python
model.createModel()
```

```python
batchSize = 128
epochs = 300
loss = "binary_crossentropy"
optimizer = "adam"
metric = "accuracy"
patience = 10
verbose = 0
sampleSize = model.X_train.shape[0]

model.setConfig(loss, optimizer, metric, batchSize, epochs, patience, verbose)
model.setTrainingSample(sampleSize)
```

```python
model.fitModel()
```

```python
model.test()
model.visualizeTest()
```

```python
model.test()
model.visualizeDecisionBoundry()
```

# SAGA

```python
log, population = NeuroEvolution.SAGA_V0(model,
                                      populationSize=10, 
                                      reductionRate=1,
                                      pateince=0,
                                      tolerance=0,
                                      d=False,
                                      epochs=100,
                                      maxNoChange=np.inf,
                                      initilizationMax=2,
                                      verbose=1,
                                      include=False,
                                      optimizationTask='activation',
                                      evaluation='validation',
                                      timeout=np.inf)
```

```python
f = Function(log.iloc[-1]['best_solution'])
plotFunction(f)
```

```python
model.f = f
model.createModel()
model.setConfig(loss, optimizer, metric, batchSize, epochs, patience, verbose)
model.setTrainingSample(sampleSize)
```
